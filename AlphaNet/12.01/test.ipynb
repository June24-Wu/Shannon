```python
import pandas as pd
import numpy as np
import os
import shutil
from tqdm import tqdm
```

```python
data_frame = pd.read_csv(path +'data.csv')
data_frame['timestamp'].drop_duplicates
```

```python
stock_list = os.listdir(path)
time_start , time_end= pd.to_datetime(time_start) , pd.to_datetime(time_end)
train_time_list , test_time_list = [] , []

for item in time_list:
    item = pd.to_datetime(item)
    if item < time_start:
        train_time_list.append(str(item))
    if time_start< item < time_end:
        test_time_list.append(str(item))
```

```python
trainx , trainy = [] , []
for stock in tqdm(os.listdir(path)):
    one_data = np.load(path + stock ,allow_pickle=True).item()
    for time in train_time_list:
        if (time in one_data) == False:
            continue
        trainx.append(np.array(one_data[time][0].iloc[:,:-1]).T)
        trainy.append(one_data[time][1])
```
