```python
import pandas as pd 
import numpy as np
data = pd.read_csv('alphanet_codetest_daily.csv')
stock = pd.DataFrame(data.groupby('ticker')['timestamp'].count() == 487).reset_index()
merge = data.merge(stock, on='ticker', how='left')
merge = merge[merge['timestamp_y'] == True]
merge = merge[['timestamp_x','ticker','open','high','low','close','vwap','volume','amount','turnover','free_turnover']]
merge
```

```python
one_date = merge[merge['ticker'] == 1] # 只计算一个股票的“数据图片”
```

```python
one_date.shape
```

```python
def count_value(data,method):
    if method == 'corr': # 计算相关系数
        return data.corr().iloc[0,1]
    if method == 'cov': # 计算协方差
        return data.cov().iloc[0,1]
def calculate(data,strip,method):
    li = []
    for i in range(2,10):
        for j in range(i+1,11):
            sum_value = 0
            count = 0
            for time_stamp in range(0,data.shape[0] - strip):
                count += 1
                sum_value += count_value(one_date.iloc[time_stamp:time_stamp+strip,[i,j]],method)
            li.append(sum_value/count) # 取平均操作
    return li
final_li =[]
final_li.append(calculate(one_date,4,'corr'))
final_li.append(calculate(one_date,4,'cov'))
final_li
```

```python
import torch
import torch.nn as nn
# Batch Normalization
input_data = np.array(final_li).T.astype(np.float32)
bn_torch = nn.BatchNorm1d(num_features=2)
data_torch = torch.from_numpy(input_data)
bn_output_torch = bn_torch(data_torch)
print(bn_output_torch)
```
